# Análsis de ventas con NumPy

Supongamos que tienes datos de ventas de una tienda en diferentes meses y quieres analizarlos para obtener información clave, como el total de ventas, el promedio de ventas por mes y los meses con mejor desempeño.

## Paso 1. Importar NumPy y definir los datos


```python
import numpy as np

# Datos de ventas por cada mes
ventas = np.array([2300, 2500, 2600, 2100, 2900, 3100, 2800, 2700, 3000, 3300, 3500, 3700])
# Meses del año
meses = np.array(["Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio",
                  "Julio", "Agosto", "Septiembre", "Octubre", "Noviembre", "Diciembre"])
print(ventas)
print(meses)
```

## Paso 2. Análisis de ventas con Numpy

Aplicaremos funciones clave para extraer información útil

### 1. Calcular el total y promedio de ventas al año.

```python
# Obtener el total de las ventas y el promedio, imprimirlas
# Calcular total de ventas al año
total_ventas = np.sum(ventas)

# Calcular el promedio de ventas por mes (sacamos la media aritmetica)
promedio_ventas = np.mean(ventas)

print(f"Total de ventas anual: {total_ventas}")
print(f"Promedio de ventas mensual: {promedio_ventas:.2f}")

```

### 2. Identificar el mes con más y menos ventas

```python
# Mes con ventas maximas
# Identificar el mes con ventas máximas
indice_max = np.argmax(ventas)  # Índice del valor máximo (posicion)
mes_max = meses[indice_max]  # Mes correspondiente (mes)
ventas_max = ventas[indice_max]  # Valor de la venta máxima (valor con la 1ra posicion)
print(f"Mes con mayores ventas: {mes_max} ({ventas_max})")


# Mes con ventas minimas
indice_min = np.argmin(ventas)  # Índice del valor mínimo
mes_min = meses[indice_min]  # Mes correspondiente
ventas_min = ventas[indice_min]  # Valor de la venta mínima
print(f"Mes con menores ventas: {mes_min} ({ventas_min})")

```

### 3. Determinar los meses con ventas superiores al promedio

```python
# Meses con Ventas mayores al promedio
# Calcular el promedio de ventas
promedio_ventas = np.mean(ventas)

# Encontrar los meses con ventas superiores al promedio
indices_superiores = np.where(ventas > promedio_ventas)[0]  # Índices de ventas mayores al promedio
meses_superiores = meses[indices_superiores]  # Meses correspondientes

# Mostrar resultados
print(f"Promedio de ventas: {promedio_ventas:.2f}") #.2f para redondear
print("Meses con ventas superiores al promedio:", ", ".join(meses_superiores))

```

## 4. Aplicar un incremento del 10% en las ventas

```python
# Aplicar un incremento del 10% en las ventas
ventas_incrementadas = ventas * 1.10

# Mostrar resultados
print("Ventas originales:", ventas)
print("Ventas con incremento del 10%:", ventas_incrementadas)
```

## 5. Gráficar las ventas actuales y el incremento para el próximo año

```python
# importar la libreria para gráficar matplotlib
import matplotlib.pyplot as plt

# Agregar las etiquetas
# Crear la gráfica
plt.figure(figsize=(10, 6))  # Tamaño de la figura

# Graficar ventas originales
plt.plot(meses, ventas, label="Ventas Actuales", marker='o', color='b', linestyle='-', linewidth=2)

# Graficar ventas incrementadas
plt.plot(meses, ventas_incrementadas, label="Ventas con Incremento del 10%", marker='s', color='g', linestyle='--', linewidth=2)

# Títulos y etiquetas
plt.title("Comparación de Ventas Actuales y con Incremento del 10%", fontsize=14)
plt.xlabel("Meses", fontsize=12)
plt.ylabel("Ventas ($)", fontsize=12)
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()

# Mostrar la gráfica
plt.tight_layout()  # Para que los elementos no se solapen
plt.show()
```

## 7. Detectar anomalias con un gráfico de caja (Boxplot)

```python
# Importamos la libreria seaborn
import seaborn as sns
# Crear un gráfico de caja (boxplot) para detectar anomalías
plt.figure(figsize=(10, 6))

# Boxplot con Seaborn
sns.boxplot(data=ventas, color="skyblue")

# Títulos y etiquetas
plt.title("Gráfico de Caja de Ventas", fontsize=14)
plt.ylabel("Ventas ($)", fontsize=12)

# Mostrar la gráfica
plt.tight_layout()  # Ajusta el gráfico para que todo sea visible
plt.show()
```
